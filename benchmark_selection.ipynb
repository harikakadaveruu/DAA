```python
import time
import matplotlib.pyplot as plt
import random as random
import psutil

def measure_time(algorithm,array_sizes):
    times=[]
    for n in array_sizes:
        array=[random.randint(0,n) for _ in range(n)]
        start=time.time()
        algorithm(array)
        end=time.time()
        times.append(end-start)
    return times

def selection_sort(arr):
    for i in range(len(arr)):
        min_index=i
        for j in range(i+1,len(arr)):
            if arr[min_index]>arr[j]:
                min_index=j
        arr[i],arr[min_index]=arr[min_index],arr[i]

array_sizes=[50,100,200,500,900,1000]
times=measure_time(selection_sort, array_sizes)
plt.plot(array_sizes,times)
plt.xlabel("Input size (n)")
plt.ylabel("Time (s)")
plt.title("Selection Sort Benchmark")
plt.show()

system_info = {
    "CPU": psutil.cpu_count(logical=False),
    "RAM": psutil.virtual_memory().total/(1024**3),
}
print(f"System information: {system_info}")
```
