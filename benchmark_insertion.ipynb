```python
import time
import matplotlib.pyplot as plt
import random as random
import psutil

def insertion_sort(arr):
    for i in range(1,len(arr)):
        key=arr[i]
        j=i-1
        while j>=0 and arr[j]>key:
            arr[j+1]=arr[j]
            j-=1
        arr[j+1]=key
        assert is_sorted(arr[:i+1]) 

def is_sorted(arr):
    for i in range(1,len(arr)):
        if arr[i-1]>arr[i]:
            return False
    return True

def measure_time(algorithm,array_sizes):
    times=[]
    for n in array_sizes:
        array=[random.randint(0,n) for _ in range(n)]
        start=time.time()
        algorithm(array)
        end=time.time()
        times.append(end-start)
    return times

array_sizes =[10,50,100,250,400,650,800]
times=measure_time(insertion_sort, array_sizes)
plt.plot(array_sizes,times)
plt.xlabel("Input size (n)")
plt.ylabel("Time (s)")
plt.title("Insertion Sort Benchmark")
plt.show()

system_info = {
    "CPU": psutil.cpu_count(logical=False),
    "RAM": psutil.virtual_memory().total/(1024**3),
}
print(f"System information: {system_info}")
```
