```python
import time
import matplotlib.pyplot as plt
import random as random
import psutil

def bubble_sort(arr):
    for i in range(len(arr)):
        for j in range(0,len(arr)-i-1):
            if arr[j]>arr[j+1]:
                arr[j],arr[j+1]=arr[j+1],arr[j]

def measure_time(algorithm,array_sizes):
    times=[]
    for n in array_sizes:
        array=[random.randint(0,n) for _ in range(n)]
        start=time.time()
        algorithm(array)
        end=time.time()
        times.append(end-start)
    return times

array_sizes=[5,10,30,50,100,200,500,1000,2000,4000]
times=measure_time(bubble_sort,array_sizes)
plt.plot(array_sizes,times)
plt.xlabel("Input size (n)")
plt.ylabel("Time (s)")
plt.title("Bubble Sort Benchmark")
plt.show()

system_info = {
    "CPU": psutil.cpu_count(logical=False),
    "RAM": psutil.virtual_memory().total/(1024**3), 
}
print(f"System information: {system_info}")
```
